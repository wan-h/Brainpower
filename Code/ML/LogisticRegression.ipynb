```python
# 实现逻辑回归
%matplotlib inline

import numpy as np
from sklearn import linear_model, datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# 加载数据
iris = datasets.load_iris()
x = iris.data[:, :2] # 使用前两个特征
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
# 构建训练模型
clf = linear_model.LogisticRegression()
clf.fit(x_train, y_train)

pred = clf.predict(x_test)
acc = clf.score(x_test, y_test)
print(f"acc: {acc}")
```

```python
# 可视化
label_0_x = x[np.argwhere(y==0)].squeeze(1)
label_1_x = x[np.argwhere(y==1)].squeeze(1)
label_2_x = x[np.argwhere(y==2)].squeeze(1)

label_0_x_test = x_test[np.argwhere(pred==0)].squeeze(1)
label_1_x_test = x_test[np.argwhere(pred==1)].squeeze(1)
label_2_x_test = x_test[np.argwhere(pred==2)].squeeze(1)

plt.figure(figsize=(10,5))
plt.plot(label_0_x[:, 0].tolist(), label_0_x[:, 1].tolist(), 'bo', label='0')
plt.plot(label_1_x[:, 0].tolist(), label_1_x[:, 1].tolist(), 'go', label='1')
plt.plot(label_2_x[:, 0].tolist(), label_2_x[:, 1].tolist(), 'ro', label='2')
plt.scatter(label_0_x_test[:, 0].tolist(), label_0_x_test[:, 1].tolist(),  marker='o', alpha=0.3, c='b', s=150, label='t-0')
plt.scatter(label_1_x_test[:, 0].tolist(), label_1_x_test[:, 1].tolist(),  marker='o', alpha=0.3, c='g', s=150, label='t-1')
plt.scatter(label_2_x_test[:, 0].tolist(), label_2_x_test[:, 1].tolist(),  marker='o', alpha=0.3, c='r', s=150, label='t-2')
plt.legend()
```
